## Problema 3.23

For the minimum mass tubular column design problem formulated in Section 2.7, consider the 
following data: P = 50 kN; l = 5.0 m; modulus of elasticity, E= 210 GPa; allowable stress, σa= 250 
MPa; mass density, 𝜌 = 7850 kg/m3.

Treating mean radius R and wall thickness t as design variables, solve the design problem 
graphically imposing an additional constraint R/t ≤ 50. This constraint is needed to avoid local 
crippling of the column. Also impose the member size constraints as

\begin{align*}
0.01 &\le R \le 1 m\\
5 &\le t \le 200 mm
\end{align*}

---
De acordo com o observado no problema 2.7, pode-se montar o problema de otimização:

\begin{align*}
\mbox{Minimizar}\  F(R,t) &= 2\rho l \pi Rt\\
\mbox{suj. a}\  \frac{P}{2\pi R t} - \sigma_a &\le 0 \\
P - \frac{\pi^3ER^3t}{4l^2} &\le 0 \\
\frac{R}{t} - 50 &\le 0 \\
R_{min} - R &\le 0 \\
R - R_{max}&\le 0 \\
t_{min} - t &\le 0 \\
t - t_{max} -&\le 0 \\
\end{align*}


````python
x1    = np.linspace(0.001,5,500)
x2    = np.linspace(0.001,300,500)*1e-3
X1,X2 = np.meshgrid(x1,x2)

p     = 50e3
l     = 5
E     = 210e9
sigma = 250e6
rho   = 7850


f  = 2*rho*l*np.pi*X1*X2
g1 = X1/X2 - 50
g2 = p - np.pi**3*E*X1**3*X2/(4*l**2)
g3 = sigma - p/(2*np.pi*X1*X2)
g4 = -X1 + 0.01
g5 =  X1 - 1
g6 = -X2 + 5e-3
g7 =  X2 - 200e-3

plt.figure(7,figsize = (8,5))

plt. contour(X1,X2,g1,0,colors = 'k')
plt. contour(X1,X2,g2,0,colors = 'r',linewidths = 3)
plt. contour(X1,X2,g3,0,colors = 'k')
plt. contour(X1,X2,g4,0,colors = 'k')
plt. contour(X1,X2,g5,0,colors = 'k')
plt. contour(X1,X2,g6,0,colors = 'r',linewidths = 3)
plt. contour(X1,X2,g7,0,colors = 'k')
plt.fill([0.035,0.035,0.037,0.04,0.045,0.053,0.06,0.06],[0.020,0.0170,0.0145,0.0120,0.008,0.005,0.005,0.02],color = 'lightgray')

x1 = plt. contour(X1,X2,f,np.arange(10)*22.09,colors = 'k',linestyles ='dashed')
plt.clabel(x1,inline = True,fontsize = 12)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$',rotation = 0)
plt.xlim(0.035,0.06);plt.ylim(0.00,0.02)
plt.title('Resolução do problema 3.23')
plt.grid(False)
````

o mínimo da função é 66 kg, com variáveis de projeto iguais a $\vec{x} = [53.6\  5]^T mm.$ As funções $g_2(x)$ e $g_1(x)$ são as restrições ativas do problema.

## Problema 3.25

Formulate the minimum mass column design problem of Section 2.7 using a hollow square cross section with outside dimension w and thickness t as design variables. Solve the problem graphically using the constraints and the data given in Exercise 3.23.

Escrevem-se as funções de área e inércia da seção quadrada vazada por meio das equações:
\begin{align*}
A &= 4at - 4t^2 \\
I &= \frac{a^4-(a-2t)^2}{12}
\end{align*}

---
O restante da solução é idêntica às soluções dos problemas 3.24 e 3.23, podendo-se escrever o problema de minimização como:

\begin{align*}
\mbox{Minimizar}\  F(a,t) &= 2\rho l (4at - 4t^2)\\
\mbox{suj. a}\  \frac{P}{4at - 4t^2} - \sigma_a &\le 0 \\
P - \frac{\pi^3E(a^4-(a-2t)^2)}{48l^2} &\le 0 \\
\frac{a}{t} - 151 &\le 0 \\
R_{min} - \frac{a}{t}  &\le 0 \\
\frac{a}{t} - R_{max}&\le 0 \\
t_{min} - t &\le 0 \\
t - t_{max} -&\le 0 \\
\end{align*}



````python
x1    = np.linspace(0.00,0.11,500)
x2    = np.linspace(0.001,250,500)*1e-3
X1,X2 = np.meshgrid(x1,x2)

p     = 50e3
l     = 5
E     = 210e9
sigma = 250e6
rho   = 7850

A     = X1**2 -(X1-2*X2)**2
I     = (X1**4-(X1-2*X2)**4)/12


f  = rho*l*A

g1 = p/A - sigma
g2 = -np.pi**2*E*I/l**2 + p
g3 = X1/X2 - 151
g4 = 200e-3 - X1/X2
g5 = (X1-X2) - 2
g6 = -X2 + 5e-3
g7 = X2 - 200e-3
plt.figure(8,figsize = (8,5))
plt.fill([0.052,0.053,0.054,0.061,0.11,0.11],[0.025,0.012,0.010,0.005,0.005,0.025],color = 'lightgray')
plt. contour(X1,X2,g1,0,colors = 'k')
plt. contour(X1,X2,g2,0,colors = 'r',linewidths = 6)
plt. contour(X1,X2,g3,0,colors = 'k')
plt. contour(X1,X2,g4,0,colors = 'k')
plt. contour(X1,X2,g5,0,colors = 'k')
plt. contour(X1,X2,g6,0,colors = 'r',linewidths = 6)
plt. contour(X1,X2,g7,0,colors = 'k')



x1 = plt. contour(X1,X2,f,np.arange(6)*70/2,colors = 'k',linestyles ='dashed')
plt.clabel(x1,inline = True,fontsize = 12)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$',rotation = 0)
plt.xlim(0.04,0.11);plt.ylim(0.0,0.025)
plt.title('Resolução do problema 3.25')
plt.grid(False)
````

## Problema 3.28

In the design of a closed-end, thin-walled cylindrical pressure vessel shown in Fig. E3.28, the design objective is to select the mean radius R and wall thickness t to minimize the total mass. The vessel should contain at least 25.0 m3 of gas at an internal pressure of 3.5 MPa. It is required that the  circumferential stress in the pressure vessel not exceed 210 MPa and the circumferential strain not exceed (1.0E-03). The circumferential stress and strain are calculated from the equations:
\begin{align*}
\sigma_c &= \frac{PR}{t} \\
\epsilon_c &= \frac{PR(2-\nu}{2Et}
\end{align*}

---
O restante da solução é idêntica às soluções dos problemas 3.24 e 3.23, podendo-se escrever o problema de minimização como:

\begin{align*}
\mbox{Minimizar}\  F(a,t) &= 2\pi \rho l Rt\\
\mbox{suj. a}\  25 - \pi(R-t/2)^2l  &\le 0\\
\frac{PR}{t} - \sigma_a &\le 0 \\
\frac{PR(2-\nu)}{2Et} - \epsilon_c &\le 0 \\
-R  + 0.5t  &\le 0 \\
-R &\le 0 \\
-t &\le 0 \\
\end{align*}





````python
x1    = np.linspace(-0.5,2,500)
x2    = np.linspace(-0.001,0.2,500)
X1,X2 = np.meshgrid(x1,x2)

p     = 3.5e6
l     = 8
E     = 210e9
sigma = 210e6
rho   = 7850
n     = 0.3


f  = rho*l*2*np.pi*X1*X2

g1 = 25 - np.pi*(X1-X2/2)**2*l
g2 = p*X1/X2 - 210
g3 = p*X1*(2-n)/2/E/X2 - 1e-3
g4 = -X1+X2/2
g5 = -X1
g6 = -X2




plt.figure(8,figsize = (8,5))
plt.fill([1,1.1,2,2],[0.015,0.2,0.2,0.03],color = 'lightgray')
plt. contour(X1,X2,g1,0,colors = 'r',linewidths = 3)
plt. contour(X1,X2,g2,0,colors = 'k')
plt. contour(X1,X2,g3,0,colors = 'r',linewidths = 3)
plt. contour(X1,X2,g4,0,colors = 'k')
plt. contour(X1,X2,g5,0,colors = 'k')
plt. contour(X1,X2,g6,0,colors = 'k')



x1 = plt. contour(X1,X2,f,np.arange(6)*6300,colors = 'k',linestyles ='dashed')
plt.clabel(x1,inline = True,fontsize = 12)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$',rotation = 0)
plt.title('Resolução do problema 3.28')

plt.grid(False)
````

o mínimo da função é 6300 kg, com variáveis de projeto iguais a $\vec{x} = [1\  0.0168]^T m.$ As funções $g_1(x)$ e $g_2(x)$ são as restrições ativas do problema.

## Problema 3.29

Consider the symmetric three-bar truss design problem formulated in Section 2.10. Formulate and solve the problem graphically for the following data:l = 1.0 m; P = 100 kN; 𝜃 = 30°; mass density, 𝜌 = 2800 kg/m3; modulus of elasticity, E = 70 GPa; allowable stress, 𝜎𝑎 = 140 MPa; ∆u = 0.5 cm; ∆v = 0.5 cm; 𝜔𝑜 = 50 Hz; 𝛽𝛽 = 1.0; A1, A2 ≥ 2 cm2.

---
A formulação do problema está descrita em 2.10:

\begin{align*}
\mbox{Minimizar}\  F(A_1,A_2) &= \rho l (2\sqrt{2}A_1+A_2)\\
\mbox{suj. a}\  \frac{\frac{P\cos\theta}{A_1}+\frac{P\sin\theta}{A_1+\sqrt2 A_2}}{\sqrt2} - \sigma_a  &\le 0\\
\frac{\sqrt2 P \sin\theta}{A_1 +\sqrt2A_2} - \sigma_a &\le 0\\
\frac{\sqrt2 Pl \cos\theta}{A_1 E} - \Delta_u &\le 0\\
\frac{\sqrt2 Pl \sin\theta}{A_1 +\sqrt2A_2} - \sigma_a &\le 0\\
(2\pi \omega_0)^2-\frac{3EA_1}{\rho l^2 (4A_1+ \sqrt2A_2)}&\le 0 \\
-\frac{\frac{P\sin\theta}{A_1 +\sqrt2A_2} - \frac{P\cos\theta}{A_1}}{\sqrt2} - \frac{\pi^2E\beta A_1}{2l^2} &\le 0\\
2-A_1 &\le 0 \\
